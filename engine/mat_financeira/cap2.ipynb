## Conceitos Básicos

<ul>
    <br><br><li><b>Título:</b> Ativo financeiro que representa um direito de crédito ou de propriedade, emitido pelo governo, empresas ou instituições, e que pode ser negociado no mercado.
    <ul>
        <br><br><li><b>Título Público - </b> Emitidos pelo governo (ex.: Tesouro Direto no Brasil). Representam dívida pública: o investidor empresta dinheiro ao Estado (ex.: Tesouro Selic, Tesouro IPCA+, Tesouro Prefixado)
        <br><br><li><b>Título Privado - </b> Emitidos por empresas ou instituiçẽs financeiras. Usado para captar recursos para projetos ou operações (ex.: Debêntures, CDBs, CRI/CRA, Ações, LCI, LCA)
    </ul>
    <br><br><li><b>Valor nominal de um título FV:</b> é o valor do título na data de seu vencimento
    <br><br><li><b>Valor atual ou valor descontado de um título PV:</b> é o valor que um título tem em uma data que antecede ao seu vencimento. A mesma definção do capítulo um de capital atual.
    <br><br><li><b>Desconto:</b> é a quantia a ser abatida do FV de um título ou a diferença entre o FV e o PV.
</ul>



## Descontos

### Desconto Simples

#### Desconto Simples Comercial ou Por Fora $(DC)$

Consites em negociar um título com uma instituição financeira, mediante uma redução no seu valor, visando antecipar o recebimento de parte dele.

O cálculo do desconto é dado por
$$DC = FV \times i \times n$$
<br><br>
O valor líquido recebido ou valor atual comercial $VC$ é dado pela diferença do valor futuro $FV$ e o desconto $DC$
$$VC = FV - DC$$
$$VC = FV - FV \times i \times n$$
$$VC = FV (1 - i \times n)$$
